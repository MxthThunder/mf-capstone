# Deliverable D4: Performance Analytics & Risk Metrics

This notebook demonstrates the calculation of annualized return (CAGR using 252 trading days), Sharpe Ratio, Beta vs benchmark index, and Value at Risk (VaR).

```python
import os
import sqlite3
import pandas as pd
import numpy as np

db_path = os.path.join("..", "data", "db", "bluestock_mf.db")
conn = sqlite3.connect(db_path)
```

### 1. CAGR Formula & Code (using 252 trading days)

$$\text{CAGR} = \left(\frac{V_{\text{end}}}{V_{\text{begin}}}\right)^{\frac{252}{N_{\text{trading days}}}} - 1$$

```python
# Fetch NAV history for a single scheme (amfi: 119551)
nav_df = pd.read_sql_query("""
    SELECT date, nav FROM fact_nav 
    WHERE amfi_code = 119551 
    ORDER BY date
""", conn)

nav_begin = nav_df['nav'].iloc[0]
nav_end = nav_df['nav'].iloc[-1]
n_trading_days = len(nav_df) - 1

cagr = (nav_end / nav_begin) ** (252.0 / n_trading_days) - 1
print(f"Initial NAV: {nav_begin}")
print(f"Final NAV: {nav_end}")
print(f"Trading Days: {n_trading_days}")
print(f"CAGR (Annualized): {cagr * 100:.4f}%")
```
