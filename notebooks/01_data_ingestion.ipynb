# Deliverable D1 & D3: Data Ingestion

This notebook demonstrates loading the raw CSV files, displaying their profiles (shapes, columns, data types), and fetching live NAV statistics.

```python
import os
import pandas as pd
import requests

raw_dir = os.path.join("..", "data", "raw")
print("Raw data files available:", os.listdir(raw_dir))
```

### 1. Load Fund Master registry and print profiling info

```python
df_fund = pd.read_csv(os.path.join(raw_dir, "01_fund_master.csv"))
print("Shape:", df_fund.shape)
print("\nColumns and Data Types:")
print(df_fund.dtypes)
df_fund.head()
```

### 2. Check shapes of all 10 datasets

```python
csv_files = [f for f in os.listdir(raw_dir) if f.endswith('.csv') and not f.endswith('_nav.csv')]
for f in sorted(csv_files):
    df = pd.read_csv(os.path.join(raw_dir, f))
    print(f"{f}: shape={df.shape}")
```

### 3. Fetch live NAV from mfapi.in

```python
resp = requests.get("https://api.mfapi.in/mf/125497")
if resp.status_code == 200:
    data = resp.json()
    print("Meta:", data.get("meta", {}))
    print("Data points fetched:", len(data.get("data", [])))
else:
    print("API request failed with status:", resp.status_code)
```
